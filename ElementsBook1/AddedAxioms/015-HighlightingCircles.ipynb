# Euclid Book 1, Additional Axiom 15

A circle [[Def I.15]](../Definitions/015-Circle.ipynb) may be *highlighted* in its location in space.

```julia
using Euclid
```

```julia
#setup the chart space...
chart = euclid_chart(title="Euclid's Elements Book I, Added Axiom 15: Highlighting Circles",
                     xlims=(-1,1), ylims=(-1,1))

# Describe and draw the line
euclid_legend(  chart,
                [circle_outline_legend(width=0.45f0, color=:steelblue, linewidth=0.1f0),
                 circle_outline_legend(width=0.45f0, color=opacify(:pink,0.6), linewidth=1f0)],
                [L"\text{A circle}", L"\text{may be \textit{highlighted} in its location in space.}"])

A = circle(Point2f0(0), 0.5f0, color=:steelblue)
show_complete(A)

#Highlight the line in space
A_highlight = highlight(A, width=10f0, color=:pink)

# Draw the animation!
draw_animation(chart, "gifs/015-CircleHighlight.gif") do t
    animate(A_highlight, 0.5f0π, 1f0π, 1.5f0π, t)
end
```
