# Euclid Book 1, Additional Axiom 11

A plane angle [[Def I.8]](../Definitions/008-PlaneAngle.ipynb) may be *highlighted* in its location in space.

```julia
using Euclid
```

```julia
#setup the chart space...
chart = euclid_chart(title="Euclid's Elements Book I, Added Axiom 11: Highlighting Angle",
                     xlims=(-1,1), ylims=(-1,1))

# Describe and draw the line
euclid_legend(  chart,
                [acute_angle_legend(color=:steelblue),
                 acute_angle_legend(color=opacify(:pink,0.6), linewidth=5f0, draw_angle=:outline)],
                [L"\text{A plane angle}", L"\text{may be \textit{highlighted} in its location in space.}"])

A = plane_angle(Point2f(-0.25,-0.25), 0.75f0, 0.75f0, π/3f0, color=:steelblue)
show_complete(A)

#Highlight the line in space
A_highlight = highlight(A, width=10f0, color=:pink)

# Draw the animation!
draw_animation(chart, "gifs/011-AngleHighlight.gif") do t
    animate(A_highlight, 0.5f0π, 1f0π, 1.5f0π, t)
end
```
