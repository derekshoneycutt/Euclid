# Euclid Book 1, Additional Axiom 18

The intersection [[AddAxiom I.7]](007-IntersectingLines.ipynb) of two circles [[Def I.15]](../Definitions/015-Circle.ipynb) occurs in exactly 1 or 2 points.

```julia
using Euclid
```

```julia
#setup the chart space...
chart = euclid_chart(title="Euclid's Elements Book I, Added Axiom 18: Intersecting Circles",
                     xlims=(-1,1), ylims=(-1,1))

# Describe and draw 2 lines
euclid_legend(  chart,
                [[circle_outline_legend(width=0.45f0, color=:steelblue, linewidth=0.1f0, center=Point2f0(0.25,0.5)),
                  circle_outline_legend(width=0.45f0, color=:steelblue, linewidth=0.1f0, center=Point2f0(0.75,0.5))],
                  circle_legend(color=opacify(:pink,0.6), width=0.2f0)],
                [L"\text{The intersection of two circles}",
                  L"\text{  occurs in exactly 1 or 2 points.}"])

A = circle(Point2f0(-0.25, 0), 0.5f0, color=:steelblue)
B = circle(Point2f0(0.25, 0), 0.5f0, color=:steelblue)
show_complete(A)
show_complete(B)

#Highlight the line in space
AB_intersect = intersection(A, B, point_color=:pink, text_color=:pink)
AB_intersect_highlight1 = highlight(AB_intersect[1], width=0.06f0, color=:pink)
AB_intersect_highlight2 = highlight(AB_intersect[2], width=0.06f0, color=:pink)

# Draw the animation!
draw_animation(chart, "gifs/018-IntersectingCircles.gif") do t
    animate(AB_intersect[1], 0.5f0, 1f0, t, fade_start=1.75f0, fade_end=2f0)
    animate(AB_intersect[2], 0.5f0, 1f0, t, fade_start=1.75f0, fade_end=2f0)
    animate(AB_intersect_highlight1, 1.001f0, 1.25f0, 1.5f0, t)
    animate(AB_intersect_highlight2, 1.001f0, 1.25f0, 1.5f0, t)
end
```
