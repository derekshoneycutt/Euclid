# Euclid Book 1, Additional Axiom 1

A point [[Def I.1]](../Definitions/001-Point.ipynb) may be *highlighted* in its location in space.

```julia
include("../../Euclid.jl");
```

```julia
#setup the chart space...
f = Figure()
ax = Axis(f[1,1], aspect=DataAspect(), title="Euclid's Elements Book I, Added Axiom 1: Highlighting Points")
xlims!(ax, -1,1)
ylims!(ax, -1,1)


# Describe and draw the point
text!(Point2f(-0.8, 0.8), text="A point may be highlighted in its location in space.")
A = point(Point2f(0,0))
show_complete(A)

#Highlight the point
A_highlight = highlight(A, width=0.06f0)

# Draw the animation!
framerate = 24
d_s = 24
t_rots = 2
timestamps = range(0,t_rots, step=t_rots/(d_s*framerate))

display_gif(record(f, "gifs/001-PointHighlight.gif", timestamps; framerate=framerate) do t
    # Animate highlighting the point A
    animate(A_highlight, 0.5f0, 1f0, 1.5f0, t)
end)
```
