# Euclid Book 1, Additional Axiom 1

A point [[Def I.1]](../Definitions/001-Point.ipynb) may be *highlighted* in its location in space.

```julia
using Euclid
```

## 2D

```julia
#setup the chart space...
chart = euclid_chart(title="Euclid's Elements Book I, Added Axiom 1: Highlighting Points",
                     xlims=(-1,1), ylims=(-1,1))

# Draw and Describe the point
euclid_legend(  chart,
                [circle_legend(color=:steelblue),
                 circle_legend(color=opacify(:pink,0.6), width=0.3f0)],
                [L"\text{A point}",
                 L"\text{may be \textit{highlighted} in its location in space.}"])

A = point(Point2f(0,0), point_color=:steelblue, text_color=:steelblue)
show_complete(A)

# Highlight the point
A_highlight = highlight(A, width=0.06f0, color=:pink)

# Draw the animation!
draw_animation(chart, "gifs/001-PointHighlight.gif") do t
    animate(A_highlight, 0.5f0π, 1f0π, 1.5f0π, t)
end
```

## 3D

```julia
#setup the chart space...
chart = euclid_chart3(title="Euclid's Elements Book I, Added Axiom 1: Highlighting Points",
                      xlims=(-1,1), ylims=(-1,1), zlims=(-1,1))

# Draw and Describe the point
euclid_legend(  chart,
                [circle_legend(color=:steelblue),
                 circle_legend(color=opacify(:pink,0.6), width=0.3f0)],
                [L"\text{A point}",
                 L"\text{may be \textit{highlighted} in its location in space.}"])

A = point(Point3f(0,0,0), point_color=:steelblue, text_color=:steelblue)
show_complete(A)

# Highlight the point
A_highlight = highlight(A, width=0.07f0, color=:pink)

# Draw the animation!
draw_animation(chart, "gifs/001-PointHighlight-3D.gif") do t
    animate(A_highlight, 0.5f0π, 1f0π, 1.5f0π, t)
end
```
