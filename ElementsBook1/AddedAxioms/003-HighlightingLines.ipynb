# Euclid Book 1, Additional Axiom 3

A line [[Def I.2]](../Definitions/002-Line.ipynb) may be *highlighted* in its location in space.

```julia
include("../../Euclid.jl");
```

```julia
#setup the chart space...
set_theme!(theme_dark())
f = Figure()
ax = euclid_axis(f[1,1], title="Euclid's Elements Book I, Added Axiom 3: Highlighting Line")
xlims!(ax, -1,1)
ylims!(ax, -1,1)


# Describe and draw the line
axislegend(ax,
           [line_legend(color=:steelblue), line_legend(color=opacify(:pink,0.6), linewidth=5f0)],
           [L"\text{A line}", L"\text{may be \textit{highlighted} in its location in space.}"])

A = line(Point2f(-0.25,-0.25), Point2f(0.25,0.25), color=:steelblue)
show_complete(A)

#Highlight the line in space
A_highlight = highlight(A, width=10f0, color=:pink)

# Draw the animation!
framerate = 24
d_s = 24
t_rots = 2π
timestamps = range(0,t_rots, step=t_rots/(d_s*framerate))

display_gif(record(f, "gifs/003-LineHighlight.gif", timestamps; framerate=framerate) do t

    # Animate highlighting the line
    animate(A_highlight, 0.5f0π, 1f0π, 1.5f0π, t)
end)
```
