# Euclid Book 1, Additional Axiom 3

A line [[Def I.2]](../Definitions/002-Line.ipynb) may be *highlighted* in its location in space.

```julia
include("../../Euclid.jl");
```

```julia
#setup the chart space...
f = Figure()
ax = Axis(f[1,1], aspect=DataAspect(), title="Euclid's Elements Book I, Added Axiom 3: Highlighting Line")
xlims!(ax, -1,1)
ylims!(ax, -1,1)


# Describe and draw the line
text!(Point2f(-0.8, 0.8), text="A line may be highlighted in its location in space.")
A = line(Point2f(-0.25,-0.25), Point2f(0.25,0.25))
show_complete(A)

#Highlight the line in space
A_highlight = highlight(A, width=10f0)

# Draw the animation!
framerate = 24
d_s = 24
t_rots = 2
timestamps = range(0,t_rots, step=t_rots/(d_s*framerate))

display_gif(record(f, "gifs/003-LineHighlight.gif", timestamps; framerate=framerate) do t

    # Animate highlighting the line
    animate(A_highlight, 0.5f0, 1f0, 1.5f0, t)
end)
```
