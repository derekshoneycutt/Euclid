# Euclid Book 1, Additional Axiom 3

A line [[Def I.2]](../Definitions/002-Line.ipynb) may be *highlighted* in its location in space.

```julia
using Euclid
```

## 2D

```julia
#setup the chart space...
chart = euclid_chart(title="Euclid's Elements Book I, Added Axiom 3: Highlighting Line",
                     xlims=(-1,1), ylims=(-1,1))

# Describe and draw the line
euclid_legend(  chart,
                [line_legend(color=:steelblue),
                 line_legend(color=opacify(:pink,0.6), linewidth=5f0)],
                [L"\text{A line}",
                 L"\text{may be \textit{highlighted} in its location in space.}"])

A = line(Point2f(-0.25,-0.25), Point2f(0.25,0.25), color=:steelblue)
show_complete(A)

#Highlight the line in space
A_highlight = highlight(A, width=10f0, color=:pink)

# Draw the animation!
draw_animation(chart, "gifs/003-LineHighlight.gif") do t
    animate(A_highlight, 0.5f0π, 1f0π, 1.5f0π, t)
end
```
