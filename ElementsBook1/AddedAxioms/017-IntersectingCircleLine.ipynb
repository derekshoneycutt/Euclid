# Euclid Book 1, Additional Axiom 17

The intersection [[AddAxiom I.7]](007-IntersectingLines.ipynb) of a straight line [[Def I.4]](../Definitions/004-StraightLine.ipynb) and a circle [[Def I.15]](../Definitions/015-Circle.ipynb) occurs in exactly 1 or 2 points.

```julia
using Euclid
```

```julia
#setup the chart space...
chart = euclid_chart(title="Euclid's Elements Book I, Added Axiom 17: Intersecting Line and Circle",
                     xlims=(-1,1), ylims=(-1,1))

# Describe and draw 2 lines
euclid_legend(  chart,
                [[line_legend(start_y=0f0, end_y=1f0, color=:steelblue),
                  circle_outline_legend(width=0.45f0, color=:steelblue, linewidth=0.1f0)],
                  circle_legend(color=opacify(:pink,0.6), width=0.2f0)],
                [L"\text{The intersection of a straight line and a circle}",
                  L"\text{  occurs in exactly 1 or 2 points.}"])

A = circle(Point2f0(0), 0.5f0, color=:steelblue)
B = line(Point2f(-0.75,0.75), Point2f(0.75,-0.75), color=:steelblue)
show_complete(A)
show_complete(B)

#Highlight the line in space
AB_intersect = intersection(A, B, point_color=:pink, text_color=:pink)
AB_intersect_highlight1 = highlight(AB_intersect[1], width=0.06f0, color=:pink)
AB_intersect_highlight2 = highlight(AB_intersect[2], width=0.06f0, color=:pink)

# Draw the animation!
draw_animation(chart, "gifs/017-IntersectingCircleLine.gif") do t
    animate(AB_intersect[1], 0.5f0π, 1f0π, t, fade_start=1.75f0π, fade_end=2f0π)
    animate(AB_intersect[2], 0.5f0π, 1f0π, t, fade_start=1.75f0π, fade_end=2f0π)
    animate(AB_intersect_highlight1, 1.001f0π, 1.25f0π, 1.5f0π, t)
    animate(AB_intersect_highlight2, 1.001f0π, 1.25f0π, 1.5f0π, t)
end
```
