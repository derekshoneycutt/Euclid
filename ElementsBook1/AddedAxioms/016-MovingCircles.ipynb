# Euclid Book 1, Additional Axiom 16

A circle [[Def I.15]](../Definitions/015-Circle.ipynb) may be *moved* by moving the line so that all points are equal distant to a new point in space.

```julia
using Euclid
```

```julia
#setup the chart space...
chart = euclid_chart(title="Euclid's Elements Book I, Added Axiom 16: Moving Circles",
                     xlims=(-1,1), ylims=(-1,1.2))


# Describe and draw the line
euclid_legend(  chart,
                [circle_outline_legend(width=0.45f0, color=:steelblue, linewidth=0.1f0),
                 circle_legend(width=0f0)],
                [L"\text{A circle may be \textit{moved} by moving the line}",
                 L"\text{  so that all points are equal distant to a new point in space.}"])

A = circle(Point2f0(0), 0.5f0, color=:steelblue)
show_complete(A)

#Demonstrate a few moves
A_move_1 = move(A, Point2f(0.25,0.25))
A_move_2 = move(A, Point2f(-0.5,-0.5), begin_at=Point2f(0.25,0.25))
A_move_3 = move(A, Point2f(0,0), begin_at=Point2f(-0.5,-0.5))

# Draw the animation!
draw_animation(chart, "gifs/016-CircleMove.gif") do t
    animate(A_move_1, 0.75f0π, 1.0f0π, t)
    animate(A_move_2, 1.16f0π, 1.25f0π, t)
    animate(A_move_3, 1.51f0π, 1.75f0π, t)
end
```
