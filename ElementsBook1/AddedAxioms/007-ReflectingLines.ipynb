# Euclid Book 1, Additional Axiom 7

A line [[Def I.2]](../Definitions/002-Line.ipynb) may be *reflected* by moving it [[AddAxiom I.5]](005-MovingLines.ipynb) such that both extremities [[Def I.3]](../Definitions/003-LineExtremities.ipynb) are exactly opposite to their beginning position across a straight line [[Def I.4]](../Definitions/004-StraightLine.ipynb) called the *axis of reflection*.

```julia
include("../../Euclid.jl");
```

```julia
#setup the chart space...
f = Figure()
ax = Axis(f[1,1], aspect=DataAspect(), title="Euclid's Elements Book I, Added Axiom 7: Reflecting Lines")
xlims!(ax, -1,1)
ylims!(ax, -1,1.2)


# Describe and draw the line
text!(Point2f(-0.9, 1.1), text="A line may be reflected by moving it such that both extremities")
text!(Point2f(-0.8, 1.0), text="are exactly opposite to their beginning position across")
text!(Point2f(-0.8, 0.9), text="a straight line called the axis of reflection.")

lines!([Point2f(-1,0), Point2f(1,0)], color=:red, linestyle=:dot)
A = line(Point2f(-0.25,-0.25), Point2f(0.25,0.15))
show_complete(A)

# Rotate the line a couple times in each direction ()
A_reflect = reflect(A)

# Draw the animation!
framerate = 24
d_s = 24
t_rots = 2
timestamps = range(0,t_rots, step=t_rots/(d_s*framerate))

display_gif(record(f, "gifs/007-LineReflect.gif", timestamps; framerate=framerate) do t

    # Animate moving/rotating the line
    animate(A_reflect, 0.5f0, 1.5f0, t)
end)
```
