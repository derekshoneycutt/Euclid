# Euclid Book 1, Additional Axiom 4

A line [[Def I.2]](../Definitions/002-Line.ipynb) may be *moved* by moving its extremities [[Def I.3]](../Definitions/003-LineExtremities.ipynb) [[AddAxiom I.2]](002-MovingPoints.ipynb) and ensuring the line remains the same between them. A line that is not *moving* is said to be *fixed*.

```julia
include("../../Euclid.jl");
```

```julia
#setup the chart space...
f = Figure()
ax = Axis(f[1,1], aspect=DataAspect(), title="Euclid's Elements Book I, Added Axiom 4: Moving Lines")
xlims!(ax, -1,1)
ylims!(ax, -1,1.2)


# Describe and draw the line
text!(Point2f(-0.9, 1.1), text="A line may be moved by moving its extremities")
text!(Point2f(-0.8, 1.0), text="and ensuring the line remains the same between them.")
text!(Point2f(-0.9, 0.9), text="A line that is not moving is said to be fixed.")
A = line(Point2f(-0.25,-0.25), Point2f(0.25,0.25))
show_complete(A)

#Demonstrate a few moves
A_move_1 = move(A, Point2f(0,0))
A_move_2 = move(A, Point2f(-0.25,0), begin_at=Point2f(0.5,0.5), move_extremityA=false)
A_move_3 = move(A, Point2f(0.25,0.25), begin_at=Point2f(-0.25,0), move_extremityA=false)

# Draw the animation!
framerate = 24
d_s = 24
t_rots = 2
timestamps = range(0,t_rots, step=t_rots/(d_s*framerate))

display_gif(record(f, "gifs/004-LineMove.gif", timestamps; framerate=framerate) do t

    # Animate moving the line
    animate(A_move_1, 0.75f0, 1.0f0, t)
    animate(A_move_2, 1.16f0, 1.25f0, t)
    animate(A_move_3, 1.51f0, 1.75f0, t)
end)
```
