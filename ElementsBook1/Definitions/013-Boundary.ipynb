# Euclid Book 1, Definition 13

A *boundary* is that which is an extremity [[Def I.3]](003-LineExtremities.ipynb) [[Def I.6]](006-SurfaceExtremities.ipynb) of anything.

```julia
using Euclid
```

```julia
#setup the chart space...
chart = euclid_chart(title="Euclid's Elements Book I, Definition 13: Boundary",
                     xlims=(-1,1), ylims=(-1,1))

# Describe and draw the line and surface
euclid_legend(  chart,
                [[line_legend(color=:steelblue),
                  circle_legend(color=opacify(:pink,0.6), center=Point2f0(0.1,0.5)),
                  circle_legend(color=opacify(:pink,0.6), center=Point2f0(0.9,0.5))],
                 [square_legend(color=:steelblue),
                  line_legend(color=:pink, start_y=0f0, end_y=0f0, linewidth=2f0),
                  line_legend(color=:pink, start_y=1f0, end_y=1f0, linewidth=2f0),
                  vline_legend(0f0, color=:pink, start_y=1f0, end_y=0f0, linewidth=2f0),
                  vline_legend(1f0, color=:pink, start_y=1f0, end_y=0f0, linewidth=2f0)]],
                [L"\text{A \textit{boundary} is that which}",
                 L"\text{is an extremity of anything.}"])

A = line(Point2f(-0.9,-0.25), Point2f(-0.15,0.25), color=:steelblue)
show_complete(A)
B = plane_surface([Point2f0(0.15, -0.375), Point2f0(0.9, -0.375), Point2f0(0.9, 0.375), Point2f0(0.15, 0.375)],
                 color=:steelblue)
show_complete(B)

# Highlight the extremity points of the line and surface
A_highlight_extrems = highlight_extremities(A, point_color=:pink, text_color=:pink)
B_highlight_extrems = highlight_extremities(B, color=:pink, width=9f0)

# Draw the animation!
draw_animation(chart, "gifs/013-Boundary.gif") do t
    animate(A_highlight_extrems, 0.5f0π, 1f0π, 1.5f0π, t)
    animate(B_highlight_extrems, 0.5f0π, 1f0π, 1.5f0π, t)
end
```
