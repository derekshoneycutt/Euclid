# Euclid Book 1, Definition 19b

*Rectilineal figures* are those which are contained by straight lines [[Def I.4]](004-StraightLine.ipynb), ... *quadrilateral* those contained by four

```julia
using Euclid;
```

## 2D

```julia
# Actors/Figures
A = euclidean_point(-0.6f0, -0.8f0)
B = euclidean_point(-0.6f0, 0.4f0)
C = euclidean_point(0.6f0, 0.4f0)
D = euclidean_point(0.6f0, -0.8f0)
ABCD = euclidean_surface([A, B, C, D], opacity=1f0, color=:steelblue)
ABCD_surface = surface("ABCD", ABCD)
AB, BC, CD, AD = extremities(ABCD_surface, ["AB", "BC", "CD", "AD"], color=:steelblue, opacity=1f0)

# Transformations
ABCD_toshift = color_shift(:steelblue, :palevioletred1)
AB_highlight = highlight(AB, ABCD_toshift, 0.5f0π, 2f0π, add_size=10f0)
BC_highlight = highlight(BC, ABCD_toshift, 0.5f0π, 2f0π, add_size=10f0)
CD_highlight = highlight(CD, ABCD_toshift, 0.5f0π, 2f0π, add_size=10f0)
AD_highlight = highlight(AD, ABCD_toshift, 0.5f0π, 2f0π, add_size=10f0)

# Draw the animation!
chart = euclid_chart(
    title="Euclid's Elements Book I, Definition 19b: Quadrilateral Figures",
    xlims=(-1,1), ylims=(-1,1))
euclid_legend(chart,
    [square_legend(color=:steelblue),
     circle_legend(width=0f0),circle_legend(width=0f0)],
    [L"\text{\textit{Rectilineal figures} are those which are contained}",
     L"\text{    by straight lines, ... \textit{quadrilateral} those}",
     L"\text{    contained by four}"])

draw_animated_transforms(chart, "gifs/019b-Quadrilateral.gif",
    [ABCD_surface, AB, BC, CD, AD],
    [AB_highlight..., BC_highlight..., CD_highlight..., AD_highlight...],
    duration=6)
```

## 3D

```julia
# Actors/Figures
A = euclidean_point(-0.6f0, -0.8f0, 0f0)
B = euclidean_point(-0.6f0, 0.4f0, 0f0)
C = euclidean_point(0.6f0, 0.4f0, 0f0)
D = euclidean_point(0.6f0, -0.8f0, 0f0)
ABCD = euclidean_surface([A, B, C, D], opacity=1f0, color=:steelblue)
ABCD_surface = surface("ABCD", ABCD)
AB, BC, CD, AD = extremities(ABCD_surface, ["AB", "BC", "CD", "AD"], color=:steelblue, opacity=1f0)

# Transformations
ABCD_toshift = color_shift(:steelblue, :palevioletred1)
AB_highlight = highlight(AB, ABCD_toshift, 0.5f0π, 2f0π, add_size=0.03f0)
BC_highlight = highlight(BC, ABCD_toshift, 0.5f0π, 2f0π, add_size=0.03f0)
CD_highlight = highlight(CD, ABCD_toshift, 0.5f0π, 2f0π, add_size=0.03f0)
AD_highlight = highlight(AD, ABCD_toshift, 0.5f0π, 2f0π, add_size=0.03f0)

# Draw the animation!
chart = euclid_chart3xy(
    title="Euclid's Elements Book I, Definition 19b: Quadrilateral Figures",
    xlims=(-1,1), ylims=(-1,1), zlims=(-1,1))
euclid_legend(chart,
    [square_legend(color=:steelblue),
     circle_legend(width=0f0),circle_legend(width=0f0)],
    [L"\text{\textit{Rectilineal figures} are those which are contained}",
     L"\text{    by straight lines, ... \textit{quadrilateral} those}",
     L"\text{    contained by four}"])

draw_animated_transforms(chart, "gifs/019b-Quadrilateral-3D.gif",
    [ABCD_surface, AB, BC, CD, AD],
    [AB_highlight..., BC_highlight..., CD_highlight..., AD_highlight...],
    duration=6)
```
