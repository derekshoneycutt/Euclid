# Euclid Book 1, Definition 3

The *extremities* of a line [[Def I.2]](002-Line.ipynb) are points [[Def I.1]](001-Point.ipynb).

```julia
include("../../Euclid.jl");
```

```julia
#setup the chart space...
set_theme!(theme_dark())
f = Figure()
ax = euclid_axis(f[1,1], title="Euclid's Elements Book I, Definition 3: Line Extremities")
xlims!(ax, -1,1)
ylims!(ax, -1,1)


# Describe and draw the line
axislegend(ax,
           [[line_legend(color=:steelblue),
             circle_legend(color=opacify(:pink,0.6), center=Point2f0(0.1,0.5)),
             circle_legend(color=opacify(:pink,0.6), center=Point2f0(0.9,0.5))]],
           [L"\text{The \textit{extremities} of a line are points.}"])

A = line(Point2f(-0.25,-0.25), Point2f(0.25,0.25), color=:steelblue)
show_complete(A)

# Highlight the extremity points of the line
A_highlight_extrems = highlight_extremities(A, point_color=:pink, text_color=:pink)

# Draw the animation!
framerate = 24
d_s = 24
t_rots = 2π
timestamps = range(0,t_rots, step=t_rots/(d_s*framerate))

display_gif(record(f, "gifs/003-LineExtremities.gif", timestamps; framerate=framerate) do t

    # Animate highlighting the line
    animate(A_highlight_extrems, 0.5f0π, 1f0π, 1.5f0π, t)
end)
```
