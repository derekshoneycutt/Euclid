# Euclid Book 1, Definition 4

A *straight line* is a line [[Def I.2]](002-Line.ipynb) which lies evenly with the points [[Def I.1]](001-Point.ipynb) on itself.

```julia
using Euclid
using GLMakie
```

```julia
#setup the chart space...
set_theme!(theme_dark())
f = Figure()
ax = euclid_axis(f[1,1], title="Euclid's Elements Book I, Definition 4: Straight Line")
xlims!(ax, -1,1)
ylims!(ax, -1,1)


# Describe and draw the line
axislegend(ax,
           [line_legend(color=:steelblue), line_legend(linestyle=:dot, color=:pink)],
           [L"\text{A \textit{straight line} is a line which lies evenly with}",
            L"\text{the points on itself.}"])

A = line(Point2f(-0.25,-0.25), Point2f(0.25,0.25), color=:steelblue)
show_complete(A)

# Highlight that it's a straight line
A_straight = highlight_straight(A, 5, color=:pink)

# Draw the animation!
framerate = 24
d_s = 24
t_rots = 2π
timestamps = range(0,t_rots, step=t_rots/(d_s*framerate))

display_gif(record(f, "gifs/004-StraightLine.gif", timestamps; framerate=framerate) do t

    # animate sraight line highlight
    animate(A_straight, 0f0π, 2f0π, t)
end)
```
