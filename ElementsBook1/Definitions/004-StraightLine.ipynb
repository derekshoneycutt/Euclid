# Euclid Book 1, Definition 4

A *straight line* is a line [[Def I.2]](002-Line.ipynb) which lies evenly with the points [[Def I.1]](001-Point.ipynb) on itself.

```julia
using Euclid
using EuclidGLMakie
```

## 2D

```julia
# Actors/Figures
AB = line("AB", euclidean_line(Point2f(-0.25), Point2f(0.25),
    color=:steelblue, width=1f0, opacity=1f0))
AB_straight = straight_line(AB, 5, color=:palevioletred1)

# Transformations
AB_straight_highlight = highlight(AB_straight, 0.0125f0, 0.5f0π, 2f0π)

# Draw the animation!
chart = euclid_chart(
    title="Euclid's Elements Book I, Definition 4: Straight Line",
    xlims=(-1,1), ylims=(-1,1))
euclid_legend(chart,
    [line_legend(color=:steelblue), line_legend(linestyle=:dot, color=:palevioletred1)],
    [L"\text{A \textit{straight line} is a line which lies evenly with}",
     L"\text{the points on itself.}"])

draw_animated_transforms(chart, "gifs/004-StraightLine.gif",
    [AB, AB_straight], AB_straight_highlight, duration=6)
```

## 3D

```julia
# Actors/Figures
AB = line("AB", euclidean_line(Point3f(-0.25), Point3f(0.25),
    color=:steelblue, width=0.01f0, opacity=1f0))
AB_straight = straight_line(AB, 5, color=:palevioletred1)

# Transformations
AB_straight_highlight = highlight(AB_straight, 0.02f0, 0.5f0π, 2f0π)

# Draw the animation!
chart = euclid_chart3(
    title="Euclid's Elements Book I, Definition 4: Straight Line",
    xlims=(-1,1), ylims=(-1,1))
euclid_legend(chart,
    [line_legend(color=:steelblue), line_legend(linestyle=:dot, color=:palevioletred1)],
    [L"\text{A \textit{straight line} is a line which lies evenly with}",
     L"\text{the points on itself.}"])

draw_animated_transforms(chart, "gifs/004-StraightLine-3D.gif",
    [AB, AB_straight], AB_straight_highlight, duration=6)
```
