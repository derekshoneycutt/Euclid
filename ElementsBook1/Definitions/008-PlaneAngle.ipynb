# Euclid Book 1, Definition 8

A *plane angle* is the inclination to one another of two lines [[Def I.2]](002-Line.ipynb) in a plane [[Def I.7]](007-PlaneSurface.ipynb) which meet one another and do not lie in a straight line [[Def I.4]](004-StraightLine.ipynb).

```julia
include("../../Euclid.jl");
```

```julia
#setup the chart space...
set_theme!(theme_dark())
f = Figure()
ax = euclid_axis(f[1,1], title="Euclid's Elements Book I, Definition 8: Plane Angle")
xlims!(ax, -1,1)
ylims!(ax, -1,1)


# Describe and draw the line
axislegend(ax,
           [[line_legend(start_y=0f0, end_y=0.65f0, color=:steelblue),
             line_legend(start_y=0f0, end_y=0f0, color=:steelblue)],
            circle_legend(width=0f0), circle_legend(width=0f0)],
           [L"\text{A \textit{plane angle} is the inclination to one another of}",
            L"\text{two lines in a plane which meet one another}",
            L"\text{and do not lie in a straight line.}"])

A = angle(Point2f(-0.25,-0.25), 0.75f0, 0.75f0, π/3f0, color=:steelblue)

# Draw the animation!
framerate = 24
d_s = 24
t_rots = 2π
timestamps = range(0,t_rots, step=t_rots/(d_s*framerate))

display_gif(record(f, "gifs/008-PlaneAngle.gif", timestamps; framerate=framerate) do t
    # Show and hide the line
    animate(A, 0.5f0π, 1f0π, t, fade_start=1.25f0π, fade_end=1.75f0π)
end)
```
