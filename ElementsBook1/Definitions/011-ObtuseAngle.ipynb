# Euclid Book 1, Definition 11

An *obtuse angle* is an angle [[Def I.8]](008-PlaneAngle.ipynb) greater than a right angle [[Def I.10]](019-RightPerpendicular.ipynb).

```julia
using Euclid
using EuclidGLMakie
```

## 2D

```julia
# Actors/Figures
Apoint = Point2f(-0, -0.25)
Bpoint = Point2f(0.75, -0.25)
Cpoint = Point2f(-0.5, 0.25)
ABvector = Bpoint - Apoint
ACvector = Cpoint - Apoint

A = euclidean_point(Apoint)
B = euclidean_point(Bpoint)
C = euclidean_point(Cpoint)
AB = line("AB", euclidean_line(A, B, color=:khaki3, width=1f0))
AC = line("AC", euclidean_line(A, C, color=:palevioletred1, width=1f0))
BAC = plane_angle("BAC", euclidean_angle(A, ABvector, ACvector, radius=0.25f0, width=1f0,
    color=:steelblue))

# Transformations
AB_show = reveal(AB, 1f0, 0.5f0, 1f0)
AC_show = reveal(AC, 1f0, 0.5f0, 1f0)
BAC_show = reveal(BAC, 1f0, 0.5f0, 1f0)
AB_hide = reveal(AB, -1f0, 1.5f0, 2f0)
AC_hide = reveal(AC, -1f0, 1.5f0, 2f0)
BAC_hide = reveal(BAC, -1f0, 1.5f0, 2f0)

# Draw the animation!
chart = euclid_chart(
    title="Euclid's Elements Book I, Definition 11: Obtuse Angle",
    xlims=(-1,1), ylims=(-1,1))
euclid_legend(chart,
    [obtuse_angle_legend(color=:steelblue)],
    [L"\text{An \textit{obtuse angle} is an angle greater than a right angle.}"])

draw_animated_transforms(chart, "gifs/011-ObtuseAngle.gif",
    [BAC, AB, AC], [AB_show, AC_show, BAC_show, AB_hide, AC_hide, BAC_hide], duration=6)
```

## 3D

```julia
# Actors/Figures
Apoint = Point3f(-0, -0.25, 0)
Bpoint = Point3f(0.75, -0.25, 0)
Cpoint = Point3f(-0.5, 0.25, 0)
ABvector = Bpoint - Apoint
ACvector = Cpoint - Apoint

A = euclidean_point(Apoint)
B = euclidean_point(Bpoint)
C = euclidean_point(Cpoint)
AB = line("AB", euclidean_line(A, B, color=:khaki3, width=0.01f0))
AC = line("AC", euclidean_line(A, C, color=:palevioletred1, width=0.01f0))
BAC = plane_angle("BAC", euclidean_angle(A, ABvector, ACvector, radius=0.25f0,
    width=0.0075f0, color=:steelblue))

# Transformations
AB_show = reveal(AB, 1f0, 0.5f0, 1f0)
AC_show = reveal(AC, 1f0, 0.5f0, 1f0)
BAC_show = reveal(BAC, 1f0, 0.5f0, 1f0)
AB_hide = reveal(AB, -1f0, 1.5f0, 2f0)
AC_hide = reveal(AC, -1f0, 1.5f0, 2f0)
BAC_hide = reveal(BAC, -1f0, 1.5f0, 2f0)

# Draw the animation!
chart = euclid_chart3xy(
    title="Euclid's Elements Book I, Definition 11: Obtuse Angle",
    xlims=(-1,1), ylims=(-1,1), zlims=(-1,1))
euclid_legend(chart,
    [obtuse_angle_legend(color=:steelblue)],
    [L"\text{An \textit{obtuse angle} is an angle greater than a right angle.}"])

draw_animated_transforms(chart, "gifs/011-ObtuseAngle-3D.gif",
    [BAC, AB, AC], [AB_show, AC_show, BAC_show, AB_hide, AC_hide, BAC_hide], duration=6)
```
