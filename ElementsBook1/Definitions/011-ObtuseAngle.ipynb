# Euclid Book 1, Definition 11

An *obtuse angle* is an angle [[Def I.8]](008-PlaneAngle.ipynb) greater than a right angle [[Def I.10]](019-RightPerpendicular.ipynb).

```julia
using Euclid
```

```julia
#setup the chart space...
chart = euclid_chart(title="Euclid's Elements Book I, Definition 11: Obtuse Angle",
                     xlims=(-1,1), ylims=(-1,1))

# Describe and draw the line
euclid_legend(  chart,
                [obtuse_angle_legend(color=:steelblue)],
                [L"\text{An \textit{obtuse angle} is an angle greater than a right angle.}"])

A = plane_angle(Point2f(-0.25,-0.25), 0.75f0, 0.75f0, 2f0/3f0, color=:steelblue)
show_complete(A)

#Highlight the line in space
A_highlight = highlight(A, width=10f0, color=:pink)

# Draw the animation!
draw_animation(chart, "gifs/011-ObtuseAngle.gif") do t
    animate(A_highlight, 0.5f0, 1f0, 1.5f0, t)
end
```
