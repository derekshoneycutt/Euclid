# Euclid Book 1, Definition 1

A *point* is that which has no part.

```julia
include("../../Euclid.jl");
```

```julia
#setup the chart space...
f = Figure()
ax = Axis(f[1,1], aspect=DataAspect(), title="Euclid's Elements Book I, Definition 1: Point")
xlims!(ax, -1,1)
ylims!(ax, -1,1)


# Describe and draw a point
text!(Point2f(-0.8, 0.8), text="A point is that which has no part.")
A = point(Point2f(0,0))

# Draw the animation!
framerate = 24
d_s = 24
t_rots = 2
timestamps = range(0,t_rots, step=t_rots/(d_s*framerate))

display_gif(record(f, "gifs/001-Point.gif", timestamps; framerate=framerate) do t
    # Show and hide the point A
    animate(A, 0.5f0, 1f0, t, fade_start=1.75f0, fade_end=2f0)
end)
```
