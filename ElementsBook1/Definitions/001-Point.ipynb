# Euclid Book 1, Definition 1

A *point* is that which has no part.

```julia
using Euclid
using GLMakie
```

```julia
#setup the chart space...
set_theme!(theme_dark())
f = Figure()
ax = euclid_axis(f[1,1], title="Euclid's Elements Book I, Definition 1: Point")
xlims!(ax, -1,1)
ylims!(ax, -1,1)


# Draw and describe a point
axislegend(ax, [circle_legend(color=:steelblue)], [L"\text{A \textit{point} is that which has no part.}"])

A = point(Point2f(0,0), point_color=:steelblue, text_color=:steelblue)

# Draw the animation!
framerate = 24
d_s = 24
t_rots = 2π
timestamps = range(0,t_rots, step=t_rots/(d_s*framerate))

display_gif(record(f, "gifs/001-Point.gif", timestamps; framerate=framerate) do t
    # Show and hide the point A
    animate(A, 0.5f0π, 1f0π, t, fade_start=1.75f0π, fade_end=2f0π)
end)
```
