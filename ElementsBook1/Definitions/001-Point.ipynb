# Euclid Book 1, Definition 1

A *point* is that which has no part.

```julia
using Euclid
using EuclidGLMakie
```

## 2D

```julia
# Actors/Figures
A = point("A", euclidean_point(0f0, 0f0, color=:steelblue, size=0.0125f0))

# Transformations
A_show = reveal(A, 1f0, 0.5f0π, 1f0π)
A_hide = reveal(A, -1f0, 1.5f0π, 2f0π)

# Draw the animation!
chart = euclid_chart(
    title="Euclid's Elements Book I, Definition 1: Point",
    xlims=(-1,1), ylims=(-1,1))
euclid_legend(chart,
    [circle_legend(color=:steelblue)],
    [L"\text{A \textit{point} is that which has no part.}"])

draw_animated_transforms(chart, "gifs/001-Point.gif",
    [A], [A_show, A_hide], duration=6)
```

## 3D

```julia
# Actors/Figures
A = point("A", euclidean_point(0f0, 0f0, 0f0, color=:steelblue, size=0.02f0))

# Transformations
A_show = reveal(A, 1f0, 0.5f0π, 1f0π)
A_hide = reveal(A, -1f0, 1.5f0π, 2f0π)

# Draw the animation!
chart = euclid_chart3(
    title="Euclid's Elements Book I, Definition 1: Point",
    xlims=(-1,1), ylims=(-1,1), zlims=(-1,1))
euclid_legend(chart,
    [circle_legend(color=:steelblue)],
    [L"\text{A \textit{point} is that which has no part.}"])

draw_animated_transforms(chart, "gifs/001-Point-3D.gif",
    [A], [A_show, A_hide], duration=6)
```
