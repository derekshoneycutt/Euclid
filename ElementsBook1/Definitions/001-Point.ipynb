# Euclid Book 1, Definition 1

A *point* is that which has no part.

```julia
using Euclid
```

## 2D

```julia
#setup the chart space...
chart = euclid_chart(title="Euclid's Elements Book I, Definition 1: Point",
                     xlims=(-1,1), ylims=(-1,1))

# Draw and describe a point
euclid_legend(  chart,
                [circle_legend(color=:steelblue)],
                [L"\text{A \textit{point} is that which has no part.}"])

A = point(Point2f(0,0), point_color=:steelblue, text_color=:steelblue)

# Draw the animation!
draw_animation(chart, "gifs/001-Point.gif") do t
    animate(A, 0.5f0π, 1f0π, t, fade_start=1.75f0π, fade_end=2f0π)
end
```

## 3D

```julia
#setup the chart space...
chart = euclid_chart3(title="Euclid's Elements Book I, Definition 1: Point",
                      xlims=(-1,1), ylims=(-1,1), zlims=(-1,1))

# Draw and describe a point
euclid_legend(  chart,
                [circle_legend(color=:steelblue)],
                [L"\text{A \textit{point} is that which has no part.}"])

A = point(Point3f0(0,0,0), point_color=:steelblue, text_color=:steelblue)

# Draw the animation!
draw_animation(chart, "gifs/001-Point-3D.gif") do t
    animate(A, 0.5f0π, 1f0π, t, fade_start=1.75f0π, fade_end=2f0π)
end
```
