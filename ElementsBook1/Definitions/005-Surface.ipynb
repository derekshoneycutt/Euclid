# Euclid Book 1, Definition 5

A *surface* is that which has length and breadth only.

```julia
using Euclid
```

```julia
#setup the chart space...
chart = euclid_chart(title="Euclid's Elements Book I, Definition 5: Surface",
                     xlims=(-1,1), ylims=(-1,1))

# Describe and draw the surface
euclid_legend(  chart,
                [square_legend(color=:steelblue)],
                [L"\text{A \textit{surface} is that which has length and breadth only.}"])

A = plane_surface([Point2f0(-0.5, -0.5), Point2f0(0.5, -0.5), Point2f0(0.5, 0.5), Point2f0(-0.5, 0.5)],
                 color=:steelblue)

# Draw the animation!
draw_animation(chart, "gifs/005-Surface.gif") do t
    animate(A, 0.5f0π, 1f0π, t, fade_start=1.25f0π, fade_end=1.75f0π)
end
```

```julia
#setup the chart space...
chart = euclid_chart3(title="Euclid's Elements Book I, Definition 5: Surface",
                     xlims=(-1,1), ylims=(-1,1), zlims=(-1,1))

# Describe and draw the surface
euclid_legend(  chart,
                [square_legend(color=:steelblue)],
                [L"\text{A \textit{surface} is that which has length and breadth only.}"])

A = plane_surface([Point3f0(-1,-1,0),Point3f0(1,-1,0),Point3f0(1,1,0),Point3f0(-1,1,0)],
                  color=:steelblue)

# Draw the animation!
draw_animation(chart, "gifs/005-Surface-3D.gif") do t
    animate(A, 0.5f0π, 1f0π, t, fade_start=1.25f0π, fade_end=1.75f0π)
end
```
