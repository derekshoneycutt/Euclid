# Euclid Book 1, Definition 10

When a straight line [[Def I.4]](004-StraightLine.ipynb) set up on a straight line makes the adjacent angles [[Def I.8]](008-PlaneAngle.ipynb) equal to one another, each of the equal angles is *right*, and the adjacent line standing on the other is called a *perpendicular* to that on which it stands.

```julia
using Euclid
```

```julia
#setup the chart space...
chart = euclid_chart(title="Euclid's Elements Book I, Definition 10: Right Angles and Perpendiculars",
                     xlims=(-1,1), ylims=(-1,1))

# Describe and draw the line
euclid_legend(  chart,
                [[line_legend(start_y=0.25f0, end_y=0.25f0, color=:steelblue),
                  vline_legend(0.5f0, start_y=0.25f0, end_y=0.9f0, color=:skyblue)],
                 circle_legend(width=0f0), circle_legend(width=0f0)],
                [L"\text{When a straight line set up on a straight line makes the adjacent angles equal}",
                 L"\text{ to one another, each of the equal angles is \textit{right}, and the adjacent line}",
                 L"\text{standing on the other is called a \textit{perpendicular} to that on which it stands.}"])

A = plane_angle(Point2f(0f0, -0.5f0), Point2f(-0.75f0, -0.5f0), Point2f0(0f0, 0.5f0),
        color=:steelblue, linecolor=:steelblue, linecolorB=:skyblue)
B = plane_angle(Point2f(0f0, -0.5f0), Point2f(0.75f0, -0.5f0), Point2f0(0f0, 0.5f0),
        color=:steelblue, linecolor=:steelblue, linecolorB=:skyblue)
show_complete(A)
show_complete(B)

#Highlight the angles in space
A_highlight = highlight(A, width=10f0, color=:pink)
B_highlight = highlight(B, width=10f0, color=:pink)

# Draw the animation!
draw_animation(chart, "gifs/010-RightPerpendicular.gif") do t
    animate(A_highlight, 0.5f0, 1f0, 1.5f0, t)
    animate(B_highlight, 0.5f0, 1f0, 1.5f0, t)
end
```
