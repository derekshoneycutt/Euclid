# Euclid Book 1, Definition 2

A *line* is breadthless length.

```julia
using Euclid
```

## 2D

```julia
# Actors/Figures
AB = line("AB", euclidean_line(Point2f(-0.25,-0.25), Point2f(0.25,0.25),
    color=:steelblue, width=1f0))

# Transformations
AB_show = reveal(AB, 1f0, 0.5f0π, 1f0π)
AB_hide = reveal(AB, -1f0, 1.5f0π, 2f0π)

# Draw the animation!
chart = euclid_chart(
    title="Euclid's Elements Book I, Definition 2: Line",
    xlims=(-1,1), ylims=(-1,1))
euclid_legend(chart,
    [line_legend(color=:steelblue)],
    [L"\text{A \textit{line} is breadthless length.}"])

draw_animated_transforms(chart, "gifs/002-Line.gif",
    [AB], [AB_show, AB_hide], duration=6)
```

## 3D

```julia
# Actors/Figures
AB = line("AB", euclidean_line(Point3f(-0.25), Point3f(0.25),
    color=:steelblue, width=0.01f0))

# Transformations
AB_show = reveal(AB, 1f0, 0.5f0π, 1f0π)
AB_hide = reveal(AB, -1f0, 1.5f0π, 2f0π)

# Draw the animation!
chart = euclid_chart3(
    title="Euclid's Elements Book I, Definition 2: Line",
    xlims=(-1,1), ylims=(-1,1), zlims=(-1,1))
euclid_legend(chart,
    [line_legend(color=:steelblue)],
    [L"\text{A \textit{line} is breadthless length.}"])

draw_animated_transforms(chart, "gifs/002-Line-3D.gif",
    [AB], [AB_show, AB_hide], duration=6)
```
