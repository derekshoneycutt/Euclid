# Euclid Book 1, Definition 2

A *line* is breadthless length.

```julia
include("../../Euclid.jl");
```

```julia
#setup the chart space...
f = Figure()
ax = euclid_axis(f[1,1], title="Euclid's Elements Book I, Definition 2: Line")
xlims!(ax, -1,1)
ylims!(ax, -1,1)


# Describe and draw the line
axislegend(ax,
           [line_legend()],
           [L"\text{A \textit{line} is breadthless length.}"])
A = line(Point2f(-0.25,-0.25), Point2f(0.25,0.25))

# Draw the animation!
framerate = 24
d_s = 24
t_rots = 2
timestamps = range(0,t_rots, step=t_rots/(d_s*framerate))

display_gif(record(f, "gifs/002-Line.gif", timestamps; framerate=framerate) do t
    # Show and hide the line
    animate(A, 0.5f0, 1f0, t, fade_start=1.25f0, fade_end=1.75f0)
end)
```
