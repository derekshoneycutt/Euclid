# Euclid Book 1, Definition 2

A *line* is breadthless length.

```julia
using Euclid
```

## 2D

```julia
#setup the chart space...
chart = euclid_chart(title="Euclid's Elements Book I, Definition 2: Line",
                     xlims=(-1,1), ylims=(-1,1))

# Describe and draw the line
euclid_legend(  chart,
                [line_legend(color=:steelblue)],
                [L"\text{A \textit{line} is breadthless length.}"])

A = line(Point2f(-0.25,-0.25), Point2f(0.25,0.25), color=:steelblue)

# Draw the animation!
draw_animation(chart, "gifs/002-Line.gif") do t
    # Show and hide the line
    animate(A, 0.5f0, 1f0, t, fade_start=1.25f0, fade_end=1.75f0)
end
```

## 3D

```julia
#setup the chart space...
chart = euclid_chart3(title="Euclid's Elements Book I, Definition 2: Line",
                     xlims=(-1,1), ylims=(-1,1), zlims=(-1,1))

# Describe and draw the line
euclid_legend(  chart,
                [line_legend(color=:steelblue)],
                [L"\text{A \textit{line} is breadthless length.}"])

A = line(Point3f(-0.25,-0.25, -0.25), Point3f(0.25,0.25, 0.25), color=:steelblue, width=0.01f0)

# Draw the animation!
draw_animation(chart, "gifs/002-Line-3D.gif") do t
    # Show and hide the line
    animate(A, 0.5f0, 1f0, t, fade_start=1.25f0, fade_end=1.75f0)
end
```
