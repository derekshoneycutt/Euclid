# Euclid Book 1, Definition 6

The *extremities* of a surface [[Def I.5]](005-Surface.ipynb) are lines [[Def I.2]](002-Line.ipynb).

```julia
include("../../Euclid.jl");
```

```julia
#setup the chart space...
f = Figure()
ax = euclid_axis(f[1,1], title="Euclid's Elements Book I, Definition 6: Surface Extremities")
xlims!(ax, -1,1)
ylims!(ax, -1,1)


# Describe and draw the surface
axislegend(ax,
           [[square_legend(color=:steelblue),
             line_legend(color=:red, start_y=0f0, end_y=0f0, linewidth=2f0),
             line_legend(color=:red, start_y=1f0, end_y=1f0, linewidth=2f0),
             vline_legend(0f0, color=:red, start_y=1f0, end_y=0f0, linewidth=2f0),
             vline_legend(1f0, color=:red, start_y=1f0, end_y=0f0, linewidth=2f0)]],
           [L"\text{The \textit{extremities} of a surface are lines.}"])

A = just_surface([Point2f0(-0.5, -0.5), Point2f0(0.5, -0.5), Point2f0(0.5, 0.5), Point2f0(-0.5, 0.5)],
                 color=:steelblue)
show_complete(A)


# Highlight the extremity points of the surface
A_highlight_extrems = highlight_extremities(A, color=:red, width=9f0)

# Draw the animation!
framerate = 24
d_s = 24
t_rots = 2
timestamps = range(0,t_rots, step=t_rots/(d_s*framerate))

display_gif(record(f, "gifs/006-SurfaceExtremities.gif", timestamps; framerate=framerate) do t

    # Show and hide the surface
    animate(A_highlight_extrems, 0.5f0, 1f0, 1.5f0, t)
end)
```
