# Euclid Book 1, Definition 6

The *extremities* of a surface [[Def I.5]](005-Surface.ipynb) are lines [[Def I.2]](002-Line.ipynb).

```julia
using Euclid
```

```julia
#setup the chart space...
chart = euclid_chart(title="Euclid's Elements Book I, Definition 6: Surface Extremities",
                     xlims=(-1,1), ylims=(-1,1))

# Describe and draw the surface
euclid_legend(  chart,
                [[square_legend(color=:steelblue),
                  line_legend(color=:pink, start_y=0f0, end_y=0f0, linewidth=2f0),
                  line_legend(color=:pink, start_y=1f0, end_y=1f0, linewidth=2f0),
                  vline_legend(0f0, color=:pink, start_y=1f0, end_y=0f0, linewidth=2f0),
                  vline_legend(1f0, color=:pink, start_y=1f0, end_y=0f0, linewidth=2f0)]],
                [L"\text{The \textit{extremities} of a surface are lines.}"])

A = plane_surface([Point2f0(-0.5, -0.5), Point2f0(0.5, -0.5), Point2f0(0.5, 0.5), Point2f0(-0.5, 0.5)],
                 color=:steelblue)
show_complete(A)

# Highlight the extremity points of the surface
A_highlight_extrems = highlight_extremities(A, color=:pink, width=9f0)

# Draw the animation!
draw_animation(chart, "gifs/006-SurfaceExtremities.gif") do t
    animate(A_highlight_extrems, 0.5f0, 1f0, 1.5f0, t)
end
```

```julia
#setup the chart space...
chart = euclid_chart3(title="Euclid's Elements Book I, Definition 6: Surface Extremities",
                      xlims=(-1,1), ylims=(-1,1), zlims=(-1,1))

# Describe and draw the surface
euclid_legend(  chart,
                [[square_legend(color=:steelblue),
                  line_legend(color=:pink, start_y=0f0, end_y=0f0, linewidth=2f0),
                  line_legend(color=:pink, start_y=1f0, end_y=1f0, linewidth=2f0),
                  vline_legend(0f0, color=:pink, start_y=1f0, end_y=0f0, linewidth=2f0),
                  vline_legend(1f0, color=:pink, start_y=1f0, end_y=0f0, linewidth=2f0)]],
                [L"\text{The \textit{extremities} of a surface are lines.}"])

A = plane_surface(triangulate_square(Point3f0(-1,-1,0),Point3f0(1,-1,0),Point3f0(1,1,0),Point3f0(-1,1,0)),
                  color=:steelblue)
show_complete(A)

# Highlight the extremity points of the surface
A_highlight_extrems = highlight_extremities(A, [Point3f0(-1,-1,0),Point3f0(1,-1,0),Point3f0(1,1,0),Point3f0(-1,1,0)], color=:pink, width=9f0)

# Draw the animation!
draw_animation(chart, "gifs/006-SurfaceExtremities-3D.gif") do t
    animate(A_highlight_extrems, 0.5f0, 1f0, 1.5f0, t)
end
```
