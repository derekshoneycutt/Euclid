# Euclid Book 1, Definition 19

*Rectilineal figures* are those which are contained by straight lines [[Def I.4]](004-StraightLine.ipynb), *trilateral* figures being those contained by three

```julia
using Euclid
```

## 2D

```julia
# Actors/Figures
A = euclidean_point(-0.6f0, -0.8f0)
B = euclidean_point(0.6f0, -0.8f0)
C = euclidean_point(0f0, 0.23923f0)
ABC = euclidean_surface([A, B, C], opacity=1f0, color=:steelblue)
ABC_surface = surface("ABC", ABC)
AB, BC, AC = extremities(ABC_surface, ["AB", "BC", "AC"], color=:steelblue, opacity=1f0)

# Transformations
ABC_toshift = color_shift(:steelblue, :palevioletred1)
AB_highlight = highlight(AB, ABC_toshift, 0.5f0π, 2f0π, add_size=10f0)
BC_highlight = highlight(BC, ABC_toshift, 0.5f0π, 2f0π, add_size=10f0)
AC_highlight = highlight(AC, ABC_toshift, 0.5f0π, 2f0π, add_size=10f0)

# Draw the animation!
chart = euclid_chart(
    title="Euclid's Elements Book I, Definition 19: Trilateral Figures",
    xlims=(-1,1), ylims=(-1,1))
euclid_legend(chart,
    [triangle_legend(color=:steelblue),
     circle_legend(width=0f0),circle_legend(width=0f0)],
    [L"\text{\textit{Rectilineal figures} are those which are contained}",
     L"\text{    by straight lines, \textit{trilateral} figures being}",
     L"\text{    those contained by three}"])

draw_animated_transforms(chart, "gifs/019a-Trilateral.gif",
    [ABC_surface, AB, BC, AC],
    [AB_highlight..., BC_highlight..., AC_highlight...],
    duration=6)
```

## 3D

```julia
# Actors/Figures
A = euclidean_point(-0.6f0, -0.8f0, 0f0)
B = euclidean_point(0.6f0, -0.8f0, 0f0)
C = euclidean_point(0f0, 0.23923f0, 0f0)
ABC = euclidean_surface([A, B, C], opacity=1f0, color=:steelblue)
ABC_surface = surface("ABC", ABC)
AB, BC, AC = extremities(ABC_surface, ["AB", "BC", "AC"], color=:steelblue, opacity=1f0)


# Transformations
ABC_toshift = color_shift(:steelblue, :palevioletred1)
AB_highlight = highlight(AB, ABC_toshift, 0.5f0π, 2f0π, add_size=0.03f0)
BC_highlight = highlight(BC, ABC_toshift, 0.5f0π, 2f0π, add_size=0.03f0)
AC_highlight = highlight(AC, ABC_toshift, 0.5f0π, 2f0π, add_size=0.03f0)

# Draw the animation!
chart = euclid_chart3xy(
    title="Euclid's Elements Book I, Definition 19: Trilateral Figures",
    xlims=(-1,1), ylims=(-1,1), zlims=(-1,1))
euclid_legend(chart,
    [triangle_legend(color=:steelblue),
     circle_legend(width=0f0),circle_legend(width=0f0),
     circle_legend(width=0f0),circle_legend(width=0f0)],
    [L"\text{\textit{Rectilineal figures} are those which are contained}",
     L"\text{    by straight lines, \textit{trilateral} figures being}",
     L"\text{    those contained by three, \textit{quadrilateral} those}",
     L"\text{    contained by four, and \textit{multilateral} those contained}",
     L"\text{    by more than four straight lines.}"])

draw_animated_transforms(chart, "gifs/019a-Trilateral-3D.gif",
    [ABC_surface, AB, BC, AC],
    [AB_highlight..., BC_highlight..., AC_highlight...],
    duration=6)
```
