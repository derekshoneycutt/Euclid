# Euclid Book 1, Definition 16

And the point is called the *center* of the circle.

```julia
using Euclid
```

```julia
#setup the chart space...
chart = euclid_chart(title="Euclid's Elements Book I, Definition 16: Center",
                     xlims=(-1,1), ylims=(-1,1))

# Describe and draw the surface
euclid_legend(  chart,
                [circle_legend(color=:steelblue, width=0.2f0)],
                [L"\text{And the point is called the \textit{center} of the circle.}"])

A = circle(Point2f0(0), 0.5f0, color=:steelblue)
show_complete(A)

center = point(A.center, point_color=:steelblue, label="")
center_highlight = highlight(center, width=0.06f0, color=:pink)

# Draw the animation!
draw_animation(chart, "gifs/016-Center.gif") do t
    animate(center, 0.5f0π, 1f0π, t, fade_start=1.75f0π, fade_end=2f0π)
    animate(center_highlight, 1.001f0π, 1.25f0π, 1.5f0π, t)
end
```
