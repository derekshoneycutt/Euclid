# Euclid Book 1, Definition 16

And the point is called the *center* of the circle.

```julia
using Euclid
```

## 2D

```julia
# Actors/Figures
center_def = euclidean_point(Point2f0(0), color=:palevioletred1)
ABC_def = euclidean_circle(center_def, 0.5f0, endθ=2f0π, width=1f0, color=:steelblue, opacity=1f0)
ABC = circle("ABC", ABC_def)
center = point("X", center_def, showtext=false)

# Transformations
center_show = reveal(center, 1f0, 0f0π, 0.5f0π)
center_highlight = highlight(center, Point3f0(0), 0.5f0π, 2f0π, add_size=0.02f0)

# Draw the animation!
chart = euclid_chart(
    title="Euclid's Elements Book I, Definition 16: Center",
    xlims=(-1,1), ylims=(-1,1))
euclid_legend(chart,
    [circle_legend(color=:steelblue, width=0.2f0)],
    [L"\text{And the point is called the \textit{center} of the circle.}"])

draw_animated_transforms(chart, "gifs/016-Center.gif",
    [ABC, center], [center_show, center_highlight...], duration=6)
```

## 3D

```julia
# Actors/Figures
center_def = euclidean_point(Point3f0(0), color=:palevioletred1)
ABC_def = euclidean_circle(center_def, 0.5f0, endθ=2f0π, width=0.01f0, color=:steelblue, opacity=1f0)
ABC = circle("ABC", ABC_def)
center = point("X", center_def, showtext=false)

# Transformations
center_show = reveal(center, 1f0, 0f0π, 0.5f0π)
center_highlight = highlight(center, Point3f0(0), 0.5f0π, 2f0π, add_size=0.03f0)

# Draw the animation!
chart = euclid_chart3xy(
    title="Euclid's Elements Book I, Definition 16: Center",
    xlims=(-1,1), ylims=(-1,1), zlims=(-1,1))
euclid_legend(chart,
    [circle_legend(color=:steelblue, width=0.2f0)],
    [L"\text{And the point is called the \textit{center} of the circle.}"])

draw_animated_transforms(chart, "gifs/016-Center-3D.gif",
    [ABC, center], [center_show, center_highlight...], duration=6)
```
